**Дана функция f(x) = (sin(x)/2 + cos(x)/2)^2 * sin(x)

    Определить корни

    Найти интервалы, на которых функция возрастает

    Найти интервалы, на которых функция убывает

    Построить график

    Вычислить вершину

    Определить промежутки, на котором f > 0

    Определить промежутки, на котором f < 0

```python
from sympy.abc import x
from sympy import *

fx = (((sin(x)/2)+(cos(x)/2))**2*sin(x))
fx
```

**Определить корни

```python
x = Symbol('x')
root = solveset(fx, x)
root
```

```python
root = solve(fx, x)
root
```

**Найти интервалы, на которых функция возрастает

```python
# from sympy import is_decreasing
# from sympy.abc import x, y
# from sympy import S, Interval, oo
# x = Symbol('x')
# is_increasing(fx, x)

from sympy import is_increasing
from sympy.abc import x, y
from sympy import S, Interval, oo
is_increasing(x**3 - 3*x**2 + 4*x, S.Reals)
is_increasing(fx, S.Reals)


```

**Найти интервалы, на которых функция убывает

```python
from sympy import is_decreasing
from sympy.abc import x, y
from sympy import S, Interval, oo
# is_decreasing(1/(x**2 - 3*x), Interval.open(S(3)/2, 3))
is_decreasing(fx, Interval.open(-oo, 0))
```

**Построить график

```python
chart = plot(fx)
```

**Вычислить вершину

```python
top = diff(fx)
plot(top)
top = solve(top)
top
```

**Определить промежутки, на котором f > 0

```python
solve_univariate_inequality(((sin(x)/2)+(cos(x)/2))**2*sin(x)>0,x,relational=False,domain=Interval(-2*pi, 2*pi))
```

**Определить промежутки, на котором f < 0

```python
solve_univariate_inequality(fx<0,x,relational=False,domain=Interval(-2*pi, 2*pi))
```
